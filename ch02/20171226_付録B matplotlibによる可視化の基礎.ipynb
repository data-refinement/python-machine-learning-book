## B.2 描画対象のFigureの明示

- pyplotは便利なインターフェイスを提供している。１枚の図を描画する際はとても便利
- pyplotの関数を利用して描画を行う際、裏側ではFigureクラスとAxesクラスがインスタンス化されている。
- 大雑把にいうと、
    - Figureクラスはすべての描画部品を保持している
    - Axesクラスは描画に関連する大半の図形を保持、管理する
- 以下では
    - pyplotのfigure関数でFigureクラスを明示的にインスタンス化した後
    - FigureクラスのgcaメソッドによりAxesクラス（正確にはAxesSubplotクラス）をインスタンス化
    - そしてAxesSubplotクラスのplotメソッドにより100個一様乱数を折れ線グラフでプロット

```python
np.random.seed(123)
fig = plt.figure() # Figureクラスをインスタンス化
ax = fig.gca() #To get the current polar axes on the current figure:
ax.plot(np.random.rand(100))
```

```python
np.random.seed(123)
fig1, fig2 = plt.figure(1), plt.figure(2) # Figureクラスの２つのインスタンスを生成
plt.figure(1)
plt.plot(np.random.rand(100))

plt.figure(2)
plt.scatter(np.random.rand(100), np.random.rand(100))

plt.show()
```

## B.3 複数の図のプロット

### subplotsを用いる方法

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))
N = 100
np.random.seed(123)
nrow, ncol = ax.shape
print('ax.shape: ',ax.shape)
print('nrow: ', nrow)
print('ncol: ', ncol)
# 分割した各領域で乱数の散布図をプロット
for i in range(nrow):
    for j in range(ncol):
        x = np.random.rand(N)
        y = 2 * x + np.random.randn(N)
        ax[i, j].scatter(x, y)
        ax[i, j].set_title('[' + str(i) + ',' + str(j) + ']')
```

### pyplotのsubplotを用いる方法 

```python
np.random.seed(123)
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
plt.figure(figsize=(10,8))

plt.subplot(221) # 2*2分割の1枚目
plt.scatter(x, y, marker='o')

plt.subplot(222) # 2枚目
plt.scatter(x, y, marker='x')

plt.subplot(223) # 3枚目
plt.scatter(x, y, marker='x')

plt.subplot(224) # 4枚目
plt.scatter(x, y, marker='^')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

### Figureクラスのadd_subplot

```python
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(2, 2, 1)# ２行２列の1つめ（左上）
ax1.scatter(x, y)

ax2 = fig.add_subplot(2, 2, 2) # 2行2列のつめ（右上）
ax2.scatter(x, y,c = 'red')

ax3 = fig.add_subplot(223) # コンマなしで書くことも可能。（左下）
ax3.scatter(x, y, c = 'yellow')

ax4 = fig.add_subplot(224) #右下
ax4.scatter(x, y, c = 'green')
```

### B.4.1 FunctionAnimationを使う方法

```python
%matplotlib nbagg
# jupyternotebookでmatplotlibのアニメーションを作成する場合に必要
```

標準正規分布（平均=0, 標準偏差=1）に従う乱数を1,000個生成し、点の個数を変化させながらヒストグラムをプロットするアニメーションを生成する。

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

N = 1000 # フレーム数
np.random.seed(123)
x = np.random.randn(N) # データの生成

absmax_x = np.ceil(np.max(np.abs(x))) # ヒストグラムの軸の範囲
xmin, xmax = -absmax_x, absmax_x

# 描画関数
def animate(nframe):
    plt.cla() # Axesオブジェクトのクリア
    plt.hist(x[:nframe], bins=20) # ヒストグラムの描画
    plt.xlim(xmin, xmax) # 軸の範囲の設定
    plt.ylim(0, 100)
    plt.title('N=%s' % nframe)

fig = plt.figure() # Figureクラスのインスタンス化

# アニメーションのインスタンス化
#  (Figureオブジェクト、描画する関数、フレーム番号を指定)
anim = animation.FuncAnimation(fig, animate, np.arange(1, N+1))

# gifファイルに保存
# anim.save('hist.gif', writer='imagemagick', fps =20)

# アニメーションの表示
plt.show()
```

- anim.saveの行でhist.gifが作成される
- &lt; img src = "hist.gif", align = "left" &gt;と書くと、フォルダにあるgifファイルを表示できる

### B.4.2 ArtistAnimationクラスを使う方法（成功）

この方法は、imsという空のリストを用意して、imsにappendで、新たなline2Dを末尾に追加していくイメージ

#### 単純にsinカーブを描く

```python
fig = plt.figure()
ax = fig.add_subplot(111)

x = np.arange(0, 10, 0.1)
ims = []
for a in range(50):
    y = np.sin(x - a)
    im1, = ax.plot(x, y, "r") #plot関数は返り値の方がlistなので、,が必要。
    # 参考→https://qiita.com/msrks/items/e264872efa062c7d6955

    ims.append([im1]) #line2Dオブジェクトをどんどんimsに入れる

    ani = animation.ArtistAnimation(fig, ims) # ArtistAnimation(Figure,Line2D)
# ani.save("sin.gif",writer="imagemagick")
```

#### 単純にsinカーブとcosカーブを上下に描いてみる

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

x = np.arange(0, 10, 0.1)
ims = []
for a in range(50):
    y = np.sin(x - a)
    im1, = ax1.plot(x, y, "r") #plot関数は返り値の方がlistなので、,が必要。
    # 参考→https://qiita.com/msrks/items/e264872efa062c7d6955
    y = np.cos(x - a)
    im2, = ax2.plot(x, y, "r")
    ims.append([im1,im2]) #line2Dオブジェクトをどんどんimsに入れる
ani = animation.ArtistAnimation(fig, ims) # ArtistAnimation(Figure,Line2D)
# ani.save("sin_cos.gif",writer="imagemagick")
```

#### 同時に描いたり、タイトルを描いたりする

```python
fig = plt.figure() # figオブジェクトを作って
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
x = np.arange(-1, 10, 0.1)
ims = []
for a in range(50):
    y = np.sin(x - a)
    im1, = ax1.plot(x+1, y, "r")
    ax1.set_title("sin curve") # ax1のタイトル
    ax1.set_xlim(0, 10)
    im2, = ax1.plot(x,y,"b")
    y = np.cos(x - a)
    
    im3, = ax2.plot(x, y, "r")
    ax2.set_xlim(0,10)
    ax2.set_title("cos curve")
    
    fig.tight_layout()
    
    ims.append([im1,im2,im3]) #　im1,im2,im3は、line2Dオブジェクト
ani = animation.ArtistAnimation(fig, ims) # ArtistAnimation(Figure,Line2D)
# ani.save("sin_cos2.gif",writer="imagemagick")
```
